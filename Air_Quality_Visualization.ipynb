### Cleaning Data
**Inisialisasi dataframe untuk interasi**

```python
dataframes = {
    "aotizhongxin": aotizhongxin_df,
    "changping": changping_df,
    "dingling": dingling_df,
    "dongsi": dongsi_df,
    "guanyuan": guanyuan_df,
    "gucheng": gucheng_df,
    "huairou": huairou_df,
    "nongzhanguan": nongzhanguan_df,
    "shunyi": shunyi_df,
    "tiantan": tiantan_df,
    "wanliu": wanliu_df,
    "wanshouxigong": wanshouxigong_df
}
```

**Untuk membersihkan data missing value berupa string, gunakan metode imputation, yaitu mengganti nilai null dengan nilai yang dominan.**


Simpan top values dari kolom wd ke dalam list

```python
top_wd = []
for location, df in dataframes.items():
  print(location, " : ", df['wd'].value_counts().nlargest(n=1).index[0])
  top_wd.append(df['wd'].value_counts().nlargest(n=1).index[0])
```

Ganti missing value dengan top value

```python
for location, df in dataframes.items():
  df['wd'].fillna(top_wd[list(dataframes.keys()).index(location)], inplace=True)
```

**Untuk membersihkan data berupa missing value, gunakan metode yang tepat, yaitu interpolation polynomial karena data yang diolah merupakan data timeseries.**

```python
for location, df in dataframes.items():
  df.interpolate(method='polynomial', order=1, inplace=True)
```

**Mengatasi kesalahan tipe data datetime**

```python
for location, df in dataframes.items():
  df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
  df.set_index('datetime', inplace=True)
  df.drop(['year', 'month', 'day', 'hour'], axis=1, inplace=True)
```

**Menghapus Kolom yang tidak digunakan**

```python
drop_list=['No', 'SO2', 'NO2', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'wd', 'WSPM']
for location, df in dataframes.items():
  df.drop(drop_list, axis=1, inplace=True)
```

**Mengecek hasil cleaning data**

```python
for location, df in dataframes.items():
  print("\n\n", location, "\n",df.isna().sum())
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Stasiun mana yang memiliki kadar rata-rata Karbon Monoksida (CO) tertinggi dan terendah?

**Mengekstrak data CO Mean dari air_quality_df**

```python
CO_columns = [col for col in air_quality_df.columns if col.startswith('CO_')]
mean_CO_df=air_quality_df[CO_columns].copy()
mean_CO_df=mean_CO_df.describe().loc['mean']
mean_CO_df=mean_CO_df.reset_index()
mean_CO_df=mean_CO_df.rename(columns={'index': 'station', 'mean': 'CO mean'})
mean_CO_df['station'] = mean_CO_df['station'].str.replace('CO_', '', regex=False)
```

```python
mean_CO_df.head()
```

**Mengurutkan data secara descending**

```python
sorted_mean_CO_df = mean_CO_df.sort_values(by="CO mean", ascending=False)
print(sorted_mean_CO_df.to_string(index=False))
```

**Menampilkan data dalam bentuk Bar Chart**

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors1 = ["r", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
colors2 = ["#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "g"]

sns.barplot(x="CO mean", y="station", data=sorted_mean_CO_df.head(6), palette=colors1, ax=ax[0])
ax[0].set_ylabel("Station", fontsize=13)
ax[0].set_xlabel("Rata-rata Karbon Monoksida (CO)", fontsize=13)
ax[0].set_title("Rata-Rata Karbon Monoksida (CO) Tertinggi", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=13)
ax[0].tick_params(axis ='x', labelsize=13)

sns.barplot(x="CO mean", y="station", data=sorted_mean_CO_df.tail(6), palette=colors2, ax=ax[1])
ax[1].set_ylabel("Station", fontsize=13)
ax[1].set_xlabel("Rata-rata Karbon Monoksida (CO)", fontsize=13)
ax[1].set_title("Rata-Rata Karbon Monoksida (CO) Terendah", fontsize=15)
ax[1].tick_params(axis ='y', labelsize=13)
ax[1].tick_params(axis ='x', labelsize=13)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()

plt.suptitle("Rata-Rata Karbon Monoksida (CO) Tertinggi dan Terendah", fontsize=20)
plt.show()
```

### Pertanyaan 2: Urutkan stasiun dengan kadar rata-rata Particulate Matter yang berdiameter 10 mikron atau kurang (PM10) dari terendah ke tertinggi?

**Mengekstrak data PM10 Mean dari air_quality_df**

```python
PM10_columns = [col for col in air_quality_df.columns if col.startswith('PM10_')]
mean_PM10_df=air_quality_df[PM10_columns].copy()
mean_PM10_df=mean_PM10_df.describe().loc['mean']
mean_PM10_df=mean_PM10_df.reset_index()
mean_PM10_df=mean_PM10_df.rename(columns={'index': 'station', 'mean': 'PM10 mean'})
mean_PM10_df['station'] = mean_PM10_df['station'].str.replace('PM10_', '', regex=False)
```

### Pertanyaan 3: Bagaimana kadar PM2.5 pada Stasiun Wanliu pada 12 bulan terakhir?

```python
wanliu_monthly_df = wanliu_timeseries_df.reset_index()
```

```python
wanliu_monthly_df = wanliu_monthly_df.resample(rule='M', on='datetime').agg({
    "PM2.5": "mean",
})
wanliu_monthly_df.index = wanliu_monthly_df.index.strftime('%Y %B')
wanliu_monthly_df = wanliu_monthly_df.reset_index()
```

```python
wanliu_monthly_df.info()
```

```python
wanliu_monthly_df.head()
```

```python
# Create a line chart
last_1_year = wanliu_monthly_df.tail(12)
plt.figure(figsize=(12, 5))
plt.plot(
    last_1_year['datetime'],
    last_1_year['PM2.5'],
    marker='o',
    linestyle='-',
    color='r')
plt.title('PM 2.5 Monthly Average Value at Wanliu Station (Last 12 Months)')
plt.xticks(fontsize=10, rotation=45)
plt.xlabel('Month')
plt.yticks(fontsize=10)
plt.ylabel('Average Value')
plt.grid(True, linestyle='--', linewidth=0.5, which='both', color='gray')

# Show the line chart
plt.show()
```
