### Cleaning Data
**Inisialisasi dataframe untuk interasi**

```python
dataframes = {
    "aotizhongxin": aotizhongxin_df,
    "changping": changping_df,
    "dingling": dingling_df,
    "dongsi": dongsi_df,
    "guanyuan": guanyuan_df,
    "gucheng": gucheng_df,
    "huairou": huairou_df,
    "nongzhanguan": nongzhanguan_df,
    "shunyi": shunyi_df,
    "tiantan": tiantan_df,
    "wanliu": wanliu_df,
    "wanshouxigong": wanshouxigong_df
}
```

**Mengatasi kesalahan tipe data datetime**

```python
for location, df in dataframes.items():
  df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
  df.drop(['No','year', 'month', 'day', 'hour'], axis=1, inplace=True)
  df.set_index('datetime', inplace=True)
```

**Untuk membersihkan data missing value berupa string, gunakan metode imputation, yaitu mengganti nilai null dengan nilai yang dominan.**


Simpan top values dari kolom wd ke dalam list

```python
top_wd = []
for location, df in dataframes.items():
  print(location, " : ", df['wd'].value_counts().nlargest(n=1).index[0])
  top_wd.append(df['wd'].value_counts().nlargest(n=1).index[0])
```

Ganti missing value dengan top value

```python
for location, df in dataframes.items():
  df['wd'].fillna(top_wd[list(dataframes.keys()).index(location)], inplace=True)
```

```python
for location, df in dataframes.items():
  print(location, ": ", df['wd'].isna().sum())
```

**Untuk membersihkan data berupa missing value, gunakan metode interpolation polynomial karena data yang diolah merupakan data time series.**

```python
for location, df in dataframes.items():
  df.interpolate(method='polynomial', order=1, inplace=True)
```

```python
for location, df in dataframes.items():
  print("\n", location, ":\n", df.isna().sum())
```

**Cari rata-rata Karbon Monoksida (CO) pada tiap stasiun**

```python
CO_mean_values = {}

for location, df in dataframes.items():
    description = df['CO'].describe()
    mean_value = description.loc['mean']
    CO_mean_values[location] = mean_value
```

```python
print(CO_mean_values)
```

Konversi CO_mean_values dictionary/map ke dataframe

```python
CO_mean_values_df = pd.DataFrame(list(CO_mean_values.items()), columns=["Location", "CO Mean Value"])
```

```python
print(CO_mean_values_df)
```

**Cari rata-rata PM10 pada tiap stasiun**

```python
PM10_mean_values = {}

for location, df in dataframes.items():
    description = df['PM10'].describe()
    mean_value = description.loc['mean']
    PM10_mean_values[location] = mean_value
```

```python
print(PM10_mean_values)
```

Konversi PM10_mean_values dictionary/map ke dataframe

```python
PM10_mean_values_df = pd.DataFrame(list(PM10_mean_values.items()), columns=["Location", "PM10 Mean Value"])
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Stasiun mana yang memiliki kadar rata-rata Karbon Monoksida (CO) tertinggi dan terendah?

```python
sorted_CO_mean_values_df = CO_mean_values_df.sort_values(by="CO Mean Value", ascending=False)
selected_columns = sorted_CO_mean_values_df[["Location", "CO Mean Value"]]
print(sorted_CO_mean_values_df.to_string(index=False))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12, 12))

colors = ["#068DA9", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#068DA9"]

sns.barplot(x="CO Mean Value", y="Location", data=sorted_CO_mean_values_df, palette=colors, ax=ax)
ax.set_ylabel("Station", fontsize=15)
ax.set_xlabel("Rata-rata Karbon Monoksida (CO)", fontsize=15)
ax.set_title("Stasiun dengan Rata-Rata Karbon Monoksida (CO) Tertinggi dan Terendah", fontsize=20)
ax.tick_params(axis ='y', labelsize=13)
ax.tick_params(axis ='x', labelsize=13)

plt.show()
```

### Pertanyaan 2: Urutkan stasiun dengan kadar rata-rata Particulate Matter yang berdiameter 10 mikron atau kurang (PM10) dari terendah ke tertinggi?

```python
sorted_PM10_mean_values_df = PM10_mean_values_df.sort_values(by="PM10 Mean Value", ascending=True)
selected_columns = sorted_PM10_mean_values_df[["Location", "PM10 Mean Value"]]
print(sorted_PM10_mean_values_df.to_string(index=False))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12, 12))

colors = ["#068DA9"]

sns.barplot(x="PM10 Mean Value", y="Location", data=sorted_PM10_mean_values_df, palette=colors)
ax.set_ylabel("Station", fontsize=15)
ax.set_xlabel("Rata-rata PM10", fontsize=15)
ax.set_title("Urutan Rata-Rata PM10 Perstasiun (Rendah ke Tinggi)", fontsize=20)
ax.tick_params(axis ='y', labelsize=13)
ax.tick_params(axis ='x', labelsize=13)

plt.show()
```

### Pertanyaan 3: Bagaimana kadar PM2.5 pada Stasiun Wanliu pada 1 tahun terakhir?

```python
# Create a line chart
last_1_year = monthly_mean_df.tail(12)
plt.figure(figsize=(12, 5))
plt.plot(
    last_1_year.index,
    last_1_year['PM2.5'],
    marker='o',
    linestyle='-',
    color='r')
plt.title('PM 2.5 Monthly Average Value at Wanliu Station (Last 1 Year)')
plt.xticks(fontsize=10, rotation=45)
plt.xlabel('Month')
plt.yticks(fontsize=10)
plt.ylabel('Average Value')
plt.grid(True, linestyle='--', linewidth=0.5, which='both', color='gray')

# Show the line chart
plt.show()
```
